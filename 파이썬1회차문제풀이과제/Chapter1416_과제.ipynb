[문제3]

숫자를 건넬 때 일부 자릿수를 영단어로 바꾼 카드를 건네주면 프로도는 원래 숫자를 찾는 게임입니다.

다음은 숫자의 일부 자릿수를 영단어로 바꾸는 예시입니다.

1478 → "one4seveneight"  
234567 → "23four5six7"  
10203 → "1zerotwozero3"    

이렇게 숫자의 일부 자릿수가 영단어로 바뀌어졌거나,   
혹은 바뀌지 않고 그대로인 문자열 s가 주어졌을때
s가 의미하는 원래 숫자를 출력하도록 코드를 작성해주세요

s = "77three4one

---
참고로 각 숫자에 대응되는 영단어는 다음 표와 같습니다.

숫자	영단어  
0	zero  
1	one  
2	two  
3	three  
4	four  
5	five  
6	six  
7	seven  
8	eight  
9	nine  

```python
def convert_to_number(s):
    
    num_dict = {
        'zero': '0',
        'one': '1',
        'two': '2',
        'three': '3',
        'four': '4',
        'five': '5',
        'six': '6',
        'seven': '7',
        'eight': '8',
        'nine': '9'
    }
    
   
    result = ''
    i = 0
    while i < len(s):
        #
        if s[i].isdigit():
            result += s[i]
            i += 1
        else:
            
            for word in num_dict:
                if s[i:i+len(word)] == word:
                    result += num_dict[word]
                    i += len(word)
                    break
    return result


s = "77three4one"


print(convert_to_number(s))
```

예를 들어 18의 자릿수 합은 1+8=9이고, 18은 9로 나누어 떨어지므로 18은 하샤드 수입니다.

x = 17

----

입출력 예시
x	결과값  
10	true  
12	true  
11	false  
13	false  

```python
def is_harshad_number(x):
    
    y = 0
    for digit in str(x):
        y += int(digit)
    
    
    return x % y == 0


x = 17
print(is_harshad_number(x))  # 결과: False
```

[문제 3: 붕어빵 주문 처리]

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python

stock = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

bbang_type = input("붕어빵 종류를 입력해주세요: ")
bbang = int(input("붕어빵 갯수를 입력해주세요: "))


if bbang_type in stock:
    if stock[bbang_type] >= bbang:
        stock[bbang_type] -= bbang
        print(f'{bbang_type} {bbang}개 판매 완료되었습니다.')
    else:
        부족량 = bbang - stock[bbang_type]
        print(f'{bbang_type}의 재고가 부족합니다. 현재 재고는 {stock[bbang_type]}개입니다. 추가로 {부족량}개가 필요합니다.')
else:
    print("주문하신 붕어빵 종류는 저희 가게에 없습니다.")

```

[문제 4: 여러 손님 주문 처리]

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python

stock = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

while True:

    bbang_type = input("붕어빵 종류를 입력해주세요 (종료를 입력하면 종료됩니다): ")


    if bbang_type == "종료":
        break


    bbang = int(input("붕어빵 갯수를 입력해주세요: "))

   
    if bbang_type in stock:
        if stock[bbang_type] >= bbang:
            stock[bbang_type] -= bbang
            print(f'{bbang_type} {bbang}개 판매 완료되었습니다.')
        else:
            print(f'{bbang_type}의 재고가 부족합니다. 현재 재고는 {stock[bbang_type]}개입니다.')
    else:
        print("주문하신 붕어빵 종류가 없습니다.")

 
    print("현재 재고 상태:")
    for item, quantity in stock.items():
        print(f"{item}: {quantity}개")

print("주문 처리가 종료되었습니다.")
```

[문제 5: 관리자 모드]

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python

stock = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}


while True:
    bbang_type = input("붕어빵 종류를 입력해주세요 (종료를 입력하면 종료됩니다): ")

    if bbang_type == "종료":
        break

    bbang = int(input("붕어빵 갯수를 입력해주세요: "))

    if bbang_type in stock:
        if stock[bbang_type] >= bbang:
            stock[bbang_type] -= bbang
            print(f'{bbang_type} {bbang}개 판매 완료되었습니다.')
        else:
            print(f'{bbang_type}의 재고가 부족합니다. 현재 재고는 {stock[bbang_type]}개입니다.')
    else:
        print("주문하신 붕어빵 종류가 없습니다.")

    print("현재 재고 상태:")
    for item, quantity in stock.items():
        print(f"{item}: {quantity}개")

print("주문 처리가 종료되었습니다.")

while True:
    print("\n관리자 모드에 진입했습니다.")
    bbang_type = input("재고를 추가할 붕어빵 종류를 입력해주세요 (종료를 입력하면 관리자 모드 종료): ")

    if bbang_type == "종료":
        break

    if bbang_type in stock:
        추가_수량 = int(input(f"{bbang_type}의 추가할 재고 수량을 입력해주세요: "))
        stock[bbang_type] += 추가_수량
        print(f"{bbang_type}의 재고가 {추가_수량}개 추가되었습니다.")
    else:
        print("해당 붕어빵 종류는 존재하지 않습니다.")

    print("현재 재고 상태:")
    for item, quantity in stock.items():
        print(f"{item}: {quantity}개")

print("관리자 모드가 종료되었습니다.")
```

[문제 6: 매출 확인]

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python

price = {
    "팥붕어빵": 1000,
    "슈크림붕어빵": 1200,
    "초코붕어빵": 1500
}


sales = {
    "팥붕어빵": 0,
    "슈크림붕어빵": 0,
    "초코붕어빵": 0
}

sales["팥붕어빵"] = 10
sales["슈크림붕어빵"] = 5
sales["초코붕어빵"] = 3


total_revenue = 0
for item in sales:
    total_revenue += sales[item] * price[item]

print(f"총 매출은 {total_revenue}원입니다.")
```
