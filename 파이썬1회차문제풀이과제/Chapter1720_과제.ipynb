과제_1: BungeoppangShop 클래스 만들기

- __init__: 초기 재고, 가격, 판매량을 설정합니다.
- check_stock: 현재 재고를 출력합니다.
- process_order: 주문을 처리하고 재고를 업데이트합니다.
- admin_mode: 관리자 모드에서 재고를 추가합니다.
- calculate_total_sales: 총 매출을 계산하고 출력합니다.

```python
class BungeoppangShop:
    def __init__(self):
       
        self.stock = {
            "팥": 20,
            "슈크림": 20,
            "피자": 20,
            "고구마": 20
        }
        
        self.prices = {
            "팥": 1000,
            "슈크림": 1200,
            "피자": 1500,
            "고구마": 1300
        }
      
        self.total_sales = 0

    def check_stock(self):
        
        print("현재 재고 상태:")
        for bread_type, count in self.stock.items():
            print(f"{bread_type}: {count}개")

    def process_order(self, bread_type, bread_count):
        
        if bread_type not in self.stock:
            print(f"{bread_type} 붕어빵은 판매하지 않습니다.")
            return
        
        if self.stock[bread_type] >= bread_count:
            self.stock[bread_type] -= bread_count
            cost = self.prices[bread_type] * bread_count
            self.total_sales += cost
            print(f"{bread_count}개의 {bread_type} 붕어빵 주문이 완료되었습니다. 금액: {cost}원")
        else:
            print(f"재고 부족으로 {bread_type} 붕어빵 {bread_count}개를 주문할 수 없습니다.")

    def admin_mode(self):
        
        print("관리자 모드입니다.")
        action = input("재고 추가를 원하면 '추가', 아니면 '종료'를 입력하세요: ")
        
        if action == "추가":
            bread_type = input("추가할 붕어빵 종류를 입력하세요: ")
            if bread_type not in self.stock:
                print(f"{bread_type} 붕어빵은 등록되지 않은 품목입니다.")
                return
            
            additional_stock = int(input(f"추가할 {bread_type} 붕어빵의 개수를 입력하세요: "))
            self.stock[bread_type] += additional_stock
            print(f"{bread_type} 붕어빵의 재고가 {additional_stock}개 추가되었습니다.")
        else:
            print("관리자 모드를 종료합니다.")

    def calculate_total_sales(self):
        
        print(f"현재까지의 총 매출은 {self.total_sales}원입니다.")

def main():
    shop = BungeoppangShop()

    while True:
        print("\n1. 재고 확인")
        print("\n2. 붕어빵 주문")
        print("\n3. 관리자 모드")
        print("\n4. 총 매출 확인")
        print("\n5. 종료")

        choice = input("원하는 기능의 번호를 입력하세요: ")

        if choice == '1':
            shop.check_stock()
        elif choice == '2':
            bread_type = input("주문할 붕어빵 종류를 입력하세요: ")
            bread_count = int(input("주문할 붕어빵의 개수를 입력하세요: "))
            shop.process_order(bread_type, bread_count)
        elif choice == '3':
            shop.admin_mode()
        elif choice == '4':
            shop.calculate_total_sales()
        elif choice == '5':
            print("프로그램을 종료합니다.")
            break
        else:
            print("잘못된 입력입니다. 다시 시도하세요.")

if __name__ == "__main__":
    main()
```


[과제 6: 매출 계산 메서드 calculate_total_sales]

목표: calculate_total_sales 메서드를 구현하여 총 매출을 계산하고 출력합니다.

설명: self.sales와 self.prices를 사용하여 총 매출을 계산하고 출력하세요.

```python
class BungeoppangShop:
    def __init__(self):
        self.stock = {
            "팥붕어빵": 10,
            "슈크림붕어빵": 8,
            "초코붕어빵": 5
        }
        self.prices = {
            "팥붕어빵": 1000,
            "슈크림붕어빵": 1200,
            "초코붕어빵": 1500
        }
        self.sales = {
            "팥붕어빵": 0,
            "슈크림붕어빵": 0,
            "초코붕어빵": 0
        }

    def check_stock(self):
        print("현재 붕어빵 재고:")
        for bread_type, quantity in self.stock.items():
            print(f"{bread_type}: {quantity}개")

    def process_order(self, bread_type, bread_count):
        if bread_type not in self.stock:
            print(f"{bread_type} 붕어빵은 판매하지 않습니다.")
            return

        if self.stock[bread_type] >= bread_count:
            self.stock[bread_type] -= bread_count
            self.sales[bread_type] += bread_count
            total_cost = self.prices[bread_type] * bread_count
            print(f"{bread_count}개의 {bread_type} 붕어빵이 주문되었습니다. 총 금액은 {total_cost}원입니다.")
        else:
            print(f"{bread_type} 붕어빵의 재고가 부족합니다. 현재 재고: {self.stock[bread_type]}개")

    def calculate_total_sales(self):
        total_sales = 0
        for bread_type in self.sales:
            # 각 붕어빵의 총 매출 계산
            total_sales += self.sales[bread_type] * self.prices[bread_type]
        print(f"총 매출: {total_sales}원")

# 테스트
shop = BungeoppangShop()
shop.process_order("팥붕어빵", 3)
shop.process_order("슈크림붕어빵", 4)
shop.calculate_total_sales()
```

- 프로그램이 종료될 때 총 매출을 계산하고 데이터를 저장합니다.


```python
class BungeoppangShop:
    def __init__(self):
        self.stock = {
            "팥붕어빵": 10,
            "슈크림붕어빵": 8,
            "초코붕어빵": 5
        }
        self.prices = {
            "팥붕어빵": 1000,
            "슈크림붕어빵": 1200,
            "초코붕어빵": 1500
        }
        self.sales = {
            "팥붕어빵": 0,
            "슈크림붕어빵": 0,
            "초코붕어빵": 0
        }

    def check_stock(self):
        print("현재 붕어빵 재고:")
        for bread_type, quantity in self.stock.items():
            print(f"{bread_type}: {quantity}개")

    def process_order(self, bread_type, bread_count):
        if bread_type not in self.stock:
            print(f"{bread_type} 붕어빵은 판매하지 않습니다.")
            return

        if self.stock[bread_type] >= bread_count:
            self.stock[bread_type] -= bread_count
            self.sales[bread_type] += bread_count
            total_cost = self.prices[bread_type] * bread_count
            print(f"{bread_count}개의 {bread_type} 붕어빵이 주문되었습니다. 총 금액은 {total_cost}원입니다.")
        else:
            print(f"{bread_type} 붕어빵의 재고가 부족합니다. 현재 재고: {self.stock[bread_type]}개")

    def admin_mode(self):
        print("관리자 모드입니다.")
        bread_type = input("재고를 추가할 붕어빵 종류를 입력하세요: ")
        if bread_type in self.stock:
            try:
                add_count = int(input("추가할 수량을 입력하세요: "))
                self.stock[bread_type] += add_count
                print(f"{bread_type} 붕어빵 재고가 {add_count}개 추가되었습니다. 현재 재고: {self.stock[bread_type]}개")
            except ValueError:
                print("유효한 숫자를 입력하세요.")
        else:
            print(f"{bread_type} 붕어빵은 존재하지 않습니다.")

    def calculate_total_sales(self):
        total_sales = 0
        for bread_type in self.sales:
            total_sales += self.sales[bread_type] * self.prices[bread_type]
        print(f"총 매출: {total_sales}원")


def main():
    shop = BungeoppangShop()
    while True:
        print("\n모드를 선택하세요:")
        print("1. 주문 모드")
        print("2. 관리자 모드")
        print("3. 종료")
        choice = input("선택 (1, 2, 3): ")

        if choice == "1":
            bread_type = input("주문할 붕어빵 종류를 입력하세요: ")
            try:
                bread_count = int(input("주문할 수량을 입력하세요: "))
                shop.process_order(bread_type, bread_count)
            except ValueError:
                print("유효한 숫자를 입력하세요.")
        
        elif choice == "2":
            shop.admin_mode()

        elif choice == "3":
            shop.calculate_total_sales()
            print("프로그램을 종료합니다.")
            break

        else:
            print("잘못된 선택입니다. 다시 시도하세요.")

if __name__ == "__main__":
    main()
```
